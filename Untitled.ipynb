```python
#ZAMAN SERİSİ TAHMİNİ(FORECASTİNG)

data_prophet=daily_sales.rename(columns={'Date':'ds','Amount':'y'}) #sadece tarih ve tutarı alıp propthet'in istediği gibi datestamp ve y(target) yaptım
print(data_prophet)
```

ds (datestamp): Prophet, tarih sütununun isminin mutlaka $ds$ olmasını bekler
y (target): Tahmin edilmesini istediğin sayısal değerin isminin mutlaka $y$ olmasını bekler.

```python
from prophet import Prophet
model=Prophet(yearly_seasonality=False,daily_seasonality=False,weekly_seasonality=True)
model.fit(data_prophet)
```

yearly_seasonality=True (Yıllık Mevsimsellik)
Verideki aylık ve mevsimsel değişimleri incele. Geçen yılın Nisan ayındaki artış bu yıl da olabilir mi, buna bak.
weekly_seasonality=True (Haftalık Mevsimsellik)
"Pazartesi ile Pazar günü arasındaki farkı öğren. Her haftalık döngüdeki ritmi yakala."

daily_seasonality=False (Günlük Mevsimsellik) 
Prophet'te "Daily Seasonality", gün içindeki saatlik değişimleri (saat 08:00 ile 20:00 arası farklar) ifade eder.
veriyi groupby('Date') yaparak her günü tek bir satıra indirdin. Yani elinde saat bilgisi yok. * Neden False? Elinde saat bazlı veri olmadığı için model gün içindeki dalgalanmaları hesaplayamaz. Eğer bunu True yaparsan, model olmayan veriyi tahmin etmeye çalışır ve hata payın artar (overfitting riski).

```python
#gelecek 30 günü içeren dataFrame oluşturmamız lazım
future=model.make_future_dataframe(periods=30)
forecast=model.predict(future)
```

ds: Tarih
yhat: Modelin asıl tahmini
yhat_lower: Tahminin düşebileceği en düşük seviye (güven aralığı)
yhat_upper: Tahminin çıkabileceği en yüksek seviye

```python
from prophet import Prophet

# 1. Veri setini temiz bir şekilde tekrar hazırla
# Önceki denemelerden kalan hatalı sütunlar varsa temizle
data_prophet = daily_sales[['Date', 'Amount']].rename(columns={'Date': 'ds', 'Amount': 'y'})

# 2. SINIRLARI TANIMLA (Kritik Adım)
# Buradaki değerlerin tabloya yazıldığından emin oluyoruz
data_prophet['cap'] = 1500000  # Maksimum satış sınırı
data_prophet['floor'] = 0      # Sıfırın altına düşme yasağı

# 3. YENİ MODEL OLUŞTUR
# Prophet nesnesini her seferinde yeniden tanımlamak en sağlıklısıdır
m = Prophet(growth='logistic', 
            yearly_seasonality=False, # Verin 1 yıldan azsa (91 gün) bunu False yapmalısın!
            weekly_seasonality=True, 
            daily_seasonality=False)

# 4. MODELİ EĞİT
m.fit(data_prophet)

# 5. GELECEK TAKVİMİNİ OLUŞTUR VE SINIRLARI EKLE
future = m.make_future_dataframe(periods=30)
future['cap'] = 1500000
future['floor'] = 0

# 6. TAHMİN ET
forecast = m.predict(future)

# 7. SONUÇLARI KONTROL ET
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(10))
```

```python
# 1. Tahmin Grafiği (Siyah noktalar ve mavi çizgili o meşhur grafik)
fig1 = m.plot(forecast)
plt.title("Amazon Gelecek 30 Günlük Tahmin", fontsize=15)
plt.show()

# 2. Bileşenler Grafiği (Trend ve Haftalık Döngü)

fig2 = m.plot_components(forecast)
plt.show()
```

ARIMA İçin Durağanlık (Stationarity) Kontrolü 

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
def duraganlik(timeseries):
    result= adfuller(timeseries)
    print(f'ADF İst: {result[0]}')
    print("p degeri:", result[1])
    if result[1] <= 0.05:
        print("Veri durağandir.")
    else:
        print("Veri durağan değildir.Fark aliniz.")

duraganlik(daily_sales['Amount'])
```

Eğer p-değeri < 0.05 ise : Direkt ARIMA(p, 0, q) aşamasına geçelim. p=0.00000071<0.05 olduğundan verimiz duragan çıktı!!!

Eğer p-değeri > 0.05 ise: Verin durağan değil. 
Bir alt satıra daily_sales['Amount'].diff().dropna() yazarak fark alıp testi tekrar yapmalıyız.

```python
# Bu grafik Trend, Mevsimsellik ve Artıkları (Residual) ayırır
#Seasonal Decomposition(Mevsimsel Ayrışma)
parcala= seasonal_decompose(daily_sales['Amount'],model='additive',period=7)
parcala.plot()
plt.show()
```

1.grafik: Observed(Gözlemlenen) ham veridir. İniş çıkışların ADF sonucuna göre belirli bir dengededir.
Satışların 600.000 ile 1.200.000 bandı arasında gidip geldiğini görüyoruz. Grafikte çok fazla zikzak (volatilite) var. Bu da bize Amazon günlük satışlarının oldukça hareketli olduğunu ve tek başına bu grafiğe bakarak geleceği tahmin etmenin zor olduğunu söylüyor.

2.Trend: Satışların genel yönünü gösterir. Çizgi düz olmadığından yumuşak bir azalış eğilimi vardır.
Amazon satışlarımızda kısa vadeli bir yükseliş dönemi yaşanmış, ancak şu an verinin sonuna doğru hafif bir doygunluk veya azalma eğilimi (downward trend) görülüyor." Bu, stratejik olarak "Neden azalıyor?" sorusunu sormamızı sağlar.

3.Seasonal(Mevsimsellik): Düzenli birbirini tekrar eden dalgalar var o yüzden haftalık bir ritim vardır. Bu dalgaların yüksekliği sabittir o yüzden "Additive" model uygundur.
Bu, veride haftalık bir döngü (period=7) olduğunun kesin kanıtıdır. Müşteriler her hafta belirli günlerde (örneğin hafta sonları) daha çok, diğer günlerde ise standart bir şekilde daha az alışveriş yapıyor. Veriyi durağan (stationary) yapan ana güç bu düzenli döngüdür.

4.Residual(Artıklar/Hata): Trend ve mevsimellik çıktıktan sonra geriye gürültü kısmı kalır.Tamamen rastgele(beyaz gürültü gibi) görünüyo o yüzden modelimiz tüm gizli bilgileri yakalamıştır.
Eğer bu kısımdaki noktalar 0 (sıfır) çizgisinin etrafında rastgele dağılmışsa, modelimiz (Trend + Seasonal) veriyi iyi anlamış demektir.

Kritik Gözlem: Grafiğin başında ve sonunda bazı büyük sapmalar (outliers) görebilirsin. Bunlar, mevsimsellikten bağımsız gelişen özel kampanya günleri veya stok sorunları gibi "sıra dışı" olayları temsil eder.

Zaman serisi ayrıştırması (Decomposition) sonucunda, Amazon satış verilerimizin net bir haftalık mevsimselliğe sahip olduğunu gözlemledik. Trend analizi, Nisan ayındaki ivmelenmenin ardından Mayıs ayı itibarıyla hafif bir yavaşlama olduğunu gösterirken; artık (residual) değerlerinin sıfır etrafındaki rastgele dağılımı, modelimizin verideki ana yapıyı başarıyla yakaladığını kanıtlamaktadır.

```python
#Arima parametlerelerini bulalımARIMA(p, d=0, q) d=0 çünkü durağan

from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

fig,(ax1,ax2)= plt.subplots(1,2, figsize=(16,5))

plot_acf(daily_sales['Amount'],ax=ax1)
ax1.set_title('ACF Grafigi(q için)')

plot_pacf(daily_sales['Amount'],ax=ax2)
ax2.set_title('PACF Grafigi (p için)')

plt.show()
```

1. ACF (Otokorelasyon Fonksiyonu) Grafiği – q (Moving Average) Seçimi:
Çubukların mavi bölgenin dışına çıkıp sıfıra yaklaşması(decaying), verinin durağanlığının kanıtıdır. Mavi alan dışı çubuk sayısı Moving Average(MA) bileşen derecesini söyler. Belirgin taşmalar 1 veya 2 adet olduğundan q=1 veya q=2 seçimi ARIMAA modeli için başlangıç noktasıdır.
   
3. PACF (Kısmi Otokorelasyon Fonksiyonu) Grafiği – p (Autoregressive) Seçimi
Bugünkü satışın doğrudan geçmişteki belirli bir günle saf bağı ölçer.İlk çubuğun çok yüksek dışarıda olmasından sonra diğer çubukların mavi alan içinde olması (p gecikmesinden sonra sert kesilme) olarak adlandırılır. Verinin çok güçlü Autoregressive(AR) yapıya sahiptir. p=1 parametresi en doğru tercihtir.
Bugünkü amazon satışı en çok dünkü satış rakamına bağlıdır.


p (AR): PACF'den gelen 1.

d (I): ADF testimiz veriyi durağan bulduğu için fark almaya gerek yok, yani 0.

q (MA): ACF'den gelen 1 (veya model performansına göre 2).

Sonuç:en dengeli model ARIMA(1, 0, 1) modelidir.

```python
#Parametreleri bulduğuma göre modeli kurmaya geçeyim

from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

model_arima= ARIMA(daily_sales['Amount'],order=(1,0,1)) #parametreleri bulduğumuz şekilde order a girdim.
model_arima_fit= model_arima.fit()  #modeli eğitelim

print(model_arima_fit.summary()) #model özeti P>|z| 0.05ten küçükse paramtreler harika
```

const (Sabit): $0.000$ çıkmış. Yani Amazon'un günlük satışlarının belirli bir ortalaması ($863,500$) var ve bu istatistiksel olarak çok güçlü bir veri.
ar.L1 (Autoregressive): $0.000$ çıkmış. Mükemmel! Dünkü satışların bugünkü satışlar üzerinde çok güçlü bir etkisi olduğunu kanıtlıyor. ($0.79$ katsayısı, dünkü hareketin yaklaşık %80'inin bugüne taşındığını söyler).
ma.L1 (Moving Average): $0.809$ çıkmış. İşte burası ilginç! $0.05$'ten büyük olduğu için bu parametre istatistiksel olarak anlamsız. Yani geçmişteki "hata/şok" terimlerinin bugünkü tahmine doğrudan bir katkısı yok.
Yorum: Modeli sadece ARIMA(1,0,0) yani basit bir AR(1) olarak kursan bile yeterli olacakmış.
Bu durum, Amazon satışlarının geçmiş şoklardan ziyade, doğrudan bir önceki günün satış performansına dayalı bir momentumla ilerlediğini kanıtlamaktadır. Modelin yüksek Kurtosis değeri ($12.66$) ise veri setindeki kampanya dönemlerine işaret eden aykırı değerlerin varlığını doğrulamaktadır.

```python
#gelecek 30 gün için tahmin

forecast_arima= model_arima_fit.get_forecast(steps=30)
forecast_mean= forecast_arima.predicted_mean
conf_int=forecast_arima.conf_int() #güven aralıkları

arima_summary = pd.DataFrame({
    'Tahmin (yhat)': forecast_mean,
    'Alt Limit (Lower)': conf_int.iloc[:, 0],
    'Ust Limit (Upper)': conf_int.iloc[:, 1]
})

# Sadece ilk 10 günü görsek
print(arima_summary.head(10))
```

ARIMA(1,0,1) modelimizin ürettiği 10 günlük projeksiyon, satışların kısa vadede toparlanarak serinin tarihsel ortalaması olan 860.000 bandına doğru stabilize olacağını öngörüyor. Güven aralıklarının (confidence intervals) zamanla genişlemesi, pazarın dinamik yapısından kaynaklanan belirsizliği istatistiksel olarak modellerken; tahminlerin istikrarlı artışı, operasyonel planlama için 'güvenli bir liman' sunmaktadır.

```python
plt.figure(figsize=(14,7))
plt.plot(daily_sales.index,daily_sales['Amount'],label='Gerçekleşen Satışlar',color='blue') #geçmişi mavi çizgiyle çizer
#arima tahmini
forecast_index=range(len(daily_sales),len(daily_sales)+forecast_steps) #İndex ile Geçmiş verilerin bittiği yerden (örneğin 91. günden) başlayıp 121. güne kadar giden yeni bir takvim oluştururuz.
plt.plot(forecast_index, forecast_mean, label='ARIMA Tahmini',color='red',linestyle='--') #geleceği kırmızı kesik çizgiyle belirtir
plt.fill_between(forecast_index, conf_int.iloc[:,0],conf_int.iloc[:,1],color='pink',alpha=0.3,label='Güven aralığı') #tahminin etrafındaki o pembe belirsizlik bölgesini oluşturur
plt.title('ARIMA(1,0,1) Modeli ile 30 Günlük Amazon Satış Tahmini', fontsize=15)
plt.xlabel('Günler')
plt.ylabel('Satış Tutarı')
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```

Ortalamaya Dönüş (Mean Reversion): Kırmızı çizgi, tahminin başında hafif bir hareket yapıp sonra düze yakın bir hal alıyorsa, bu ARIMA'nın "uzun vadede verinin ortalamasına güvendiğini" gösterir. Prophet gibi çok agresif çıkışlar yapmaz, daha temkinlidir.

Güven Aralığı (Pembe Gölge):

Gölgenin genişliği, Amazon satışlarındaki volatilitenin (oynaklığın) bir yansımasıdır.

Tahmin süresi ilerledikçe (30. güne doğru) gölgenin genişlemesi, modelin "Uzaklaştıkça yanılma payım artıyor" uyarısıdır.

Yorum: "Satışlar pembe alanın içinde kalacak; en iyi ihtimalle üst çizgi, en kötü ihtimalle alt çizgi seviyesinde olacaktır."
